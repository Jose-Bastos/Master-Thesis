```python
dx_01_1 = np.array([53.0,55.7])
dx_01_2 = np.array([67.4,69.4])
dx_001_1 = np.array([3.12,1.62])
dx_001_2 = np.array([3.01,3.51])
dx_005_1 = np.array([1.9,1.5,2.19])
dx_005_2 = np.array([3.9,4.0,3.5])
dx_002_1 = np.array([1.19,1.3])
dx_002_2 = np.array([2.3,2.3])
dx_0001_1 = np.array([0.9,1.3])
dx_0001_2 = np.array([1.9,2.0])


dx_01 = np.array([53.0,67.43,53.0,67.30])
dx_005 = np.array([2.15,3.53,4.9,6.26,3.87,3.55])
dx_004 = np.array([1.39,2.28,1.77,1.61,2.31,2.84,1.59,2.2])
dx_001 = np.array([1.63,3.51,3.54,3.10,2.8,2.5])
dx_0005 = np.array([2.5,1.63,1.93,3.96,1.41,2.13])
dx_00002 = np.array([1.19,1.91])
dx_00001 = np.array([0.81,1.04])


mean1 = np.mean(dx_01)
std1  = np.std(dx_01)
mean2 = np.mean(dx_005)
std2  = np.std(dx_005)
mean3 = np.mean(dx_004)
std3  = np.std(dx_004)
mean4 = np.mean(dx_001)
std4  = np.std(dx_001)
mean5 = np.mean(dx_0005)
std5  = np.std(dx_0005)
mean6 = np.mean(dx_00002)
std6  = np.std(dx_00002)
mean7 = np.mean(dx_00001)
std7  = np.std(dx_00001)


means= [mean1,mean2,mean3,mean4,mean5,mean6,mean7]
stds = [std1,std2,std3,std4,std5,std6,std7]
```

```python
for i,dxs in enumerate(dx):
    print("dx is:",dxs)
    print("mean is:",means[i])
    print("std is:",stds[i])
    print()
```

```python
plt.figure(figsize=(20,10))
#xmarkers = 
plt.grid()
#plt.title("Error in coefficients as a function of spatial resolution")
plt.xlabel("Spatial resolution(m)")
plt.ylabel("Percentage of error in determined coefficients(%)")
plt.plot(dx,means,'--o',label="Error in coeficients",markersize=10)
plt.errorbar(dx,means,yerr=stds,fmt='none')
plt.savefig("spatial_res.png",format="png",bbox_inches="tight")
plt.show()
```
