```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dx = [0.1,0.01,0.005,0.002,0.001]
error1 = np.array([53.0,3.5,1.9,1.19,0.9])
error2 = np.array([67.4,6.1,3.9,2.3,1.9])
mean_error = 0.5*np.abs((error1+error2))
print(mean_error)
```

```python
plt.figure(figsize=(20,10))
#xmarkers = 
plt.grid()
plt.xlabel("Spatial resolution(m)")
plt.ylabel("Percentage of error in determined coefficients(%)")
plt.plot(dx,mean_error,'--o',label="Error in coeficients",markersize=10)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.xticks(np.linspace(0.001,0.010,10))
plt.yticks(np.linspace(0,5+5/20,20))
plt.grid()
plt.xlabel("Spatial resolution(m)")
plt.ylabel("Percentage of error in determined coefficients(%)")
plt.plot(dx[1:],mean_error[1:],'--o',label="Error in coeficients",markersize=10)
plt.show()
```
