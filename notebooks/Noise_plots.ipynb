```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})
import os
os.chdir("/notebooks")
%load_ext autoreload
%autoreload 2
```

```python
noise_levels = [0,1,5,10,25,50]
```

```python
error0  = np.array([2.0,2.2,1.9,1.93,1.9])
error1 = np.array([2.16,3.04,2.48,3.42,2.43,3.37,2.11,3.32])
error5 = np.array([	1.88,3.6,2.98,3.1,1.17,1.84])
error10 = np.array([4.96,8.53,5.32,9.89,5.13,8.35,2.63,5.07])
error25 = np.array([63.78,81.93,12.24,17.15,14.55,21.84,16.50,26.1,15.51,23.92])
error50 =np.array([51.30,74.05,51.98,80.40,74.65,84.80])
```

```python
mean0 = np.mean(error0)
mean1 = np.mean(error1)
mean5 = np.mean(error5)
mean10 = np.mean(error10)
mean25 = np.mean(error25)
mean50 = np.mean(error50)

std0 = np.std(error0)
std1 = np.std(error1)
std5 = np.std(error5)
std10 = np.std(error10)
std25 = np.std(error25)
std50= np.std(error50)

means = [mean0,mean1,mean5,mean10,mean25,mean50]
stds = [std0,std1,std5,std10,std25,std50]

for i,noise in enumerate(noise_levels):
    print("Noise level is:",noise)
    print("mean is:",means[i])
    print("std is:",stds[i])
    print()
```

```python
plt.figure(figsize=(20,10))
#xmarkers = 
plt.grid()
plt.xlabel("Noise level(m)")
plt.ylabel("Percentage of error in determined coefficients(%)")
plt.plot(noise_levels,means,'--o',label="Error in coeficients",markersize=10)
plt.errorbar(noise_levels,means,yerr=stds,fmt='none')
plt.legend()
plt.savefig("noise.png",format="png",bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(20,10))
#xmarkers = 
plt.grid()
plt.xlabel("Noise level(%)")
plt.ylabel("Percentage of error in determined coefficients(%)")
plt.hlines(3,0,10,label="y=3%",color="red",linestyle="--")
plt.xticks(np.arange(0,11,1))
plt.plot(noise_levels[:-2],means[:-2],'--o',label="Error in coeficients",markersize=10)
plt.errorbar(noise_levels[:-2],means[:-2],yerr=stds[:-2],fmt='none')
plt.legend()
plt.savefig("noise_zoom.png",format="png",bbox_inches="tight")
plt.show()
```
